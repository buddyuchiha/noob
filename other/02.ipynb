__Что такое Django? (для суперчайников (для меня))__

Запрос поступает с браузера в Django, который затем переходит в маршрутизатор urls.py



Затем он попадает в контроллер IndexViev



У нас есть в Django (БД), контроллер делает туда запрос и эта информация в виде словаря отправляется в контроллер, контроллер берет эти данные и берет Template (HTML-страница),отправляет туда наши данные из БД и соединяет их

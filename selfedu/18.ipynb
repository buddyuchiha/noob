__#18. Создание и запуск файлов миграций__

Для появления таблицы в самой БД необходимо создать файл миграции

Миграции в Django — это механизм для управления изменениями в структуре базы данных (БД) с помощью кода. Они позволяют синхронизировать модели Django (описанные в Python) с реальными таблицами в БД, отслеживая историю изменений.

В Django файлы миграции создаются автоматически, если мы в приложении каком-то определяем модель и создаем миграции, то файлы миграции появляются в папке migrations

Каждый новый файл миграции можно воспринимать как контроллер версии, мы всегда можем откатиться назад

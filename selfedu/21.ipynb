__#21. Сортировка, изменение и удаление записей__

Записи можно сортировать, используя order_by



Чтобы изменить запись, нужно:

1) Приравнять переменную к каком-либу объекту через get

2) Поменять значения

3) Сохранить все через save()



Или менять для всех через all и update:



Если для группы конкретных то нужно профильтровать через filer


Ну и удаление также, приравниваем переменную и потом delete()

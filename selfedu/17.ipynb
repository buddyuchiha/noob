Модель отвечает за хранение и оперирование данными сайта.

Официально Django поддерживает PostgreSQL, MariaDB, MySQL, Oracle, SQLite

У каждой таблицы свое число и тип столбцов (полей).

Для создания таблиц и наполнениями их данными используется язык SQL (Strucutured Query Language)

В Django редко использутеся чистый SQL.  

У каждого типа СУБД свои особенности этого языка, чтобы программист мог создавать универсальный код, непривязанный к конкретной СУБД, в Django встроен механизм ORM:



Фрейморвк сам закрывает и открывает таблицы в БД

По умолчанию Django сконфигурирован на SQLite 3



В этом словаре как раз таки и подключаются другие БДшки

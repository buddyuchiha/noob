__#11. Стандартные шаблонные фильтры__

Данные в шаблонах можно преобразовывать, используя различные фильтры.

Их вообще оч много, но ща будет разбор только некоторых:

Фильр add позволяет прибавлять некие целочисленные значения к числовому параметру



Фильтр capfirst преобразует первый символ строки в заглавную букву



Фильтр cut позволяет удалять из строки заданный символ



Можно также применить два фильтра 



Фильтр default позволяет задать значение по умолчанию, если нету переменной или она ложная

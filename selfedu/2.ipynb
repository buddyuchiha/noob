__#2. Установка Django. Создание проекта__

Django взаимодействует с другими пакетами языка Python -> чтобы все модули согласовано работали желательно использовать одни и те же версии программных продуктов. Для качественной работы используется виртуальное окружение.

Ультимативный гайд для чайников:
1) Win + R -> cmd
2) python -V (проверка что он ваще есть)
3) pip list (все установленные библиотеки)
4) создадим подкатолог, используя far manager, где будем хранить виртуальные окружения:

5) C:\django (переходим в папку)
6) python -m venv <название> (для примера djvenv)
7) .\djvenv\Scripts\activate (запускаем виртуальное окружение)
8) deactivate (выйти из виртуального окружения)



ну и в вскоде все тоже нормалек



чтобы глянуть список команд:

django-admin

Создание нового проекта:

django-admin startproject <имя проекта> (обычно в качестве имени проекта используется доменное имя сайта, для примера sitewomen)



Чтобы создать локальный сервер, вводим в терминал:

python manage.py runserver

Все нормалек:



по умолчанию Django использует бд sqlite3, но можно поменять на PostgreSQL, MySQL, Oracle

также при запуске можно указать свой порт следующей командой:

django-admin startproject sitewomen 4000

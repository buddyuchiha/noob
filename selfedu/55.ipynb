__#55. Классы CreateView и UpdateView__

CreateView используется для создания новых объектов модели.

Основные возможности:

1) Автоматически создает форму на основе модели

2) Обрабатывает валидацию данных

3) Сохраняет объект в базу данных

4) Перенаправляет после успешного создания



UpdateView используется для редактирования существующих объектов.

Основные возможности:

1) Получает объект из базы (по pk или slug из URL)

2) Создает форму с текущими значениями

3) Обрабатывает обновление данных

4) Сохраняет изменения в базу

__#52. Класс ListView__

ListView - это класс из django.views.generic, который:

1) Автоматически получает список объектов из модели (или любого другого источника данных)

2) Обрабатывает пагинацию (разбиение на страницы)

3) Рендерит шаблон с контекстом, содержащим список объектов

4) Упрощает стандартные операции для отображения списков

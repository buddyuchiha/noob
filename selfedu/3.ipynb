MTV (Models, Templates, Views) - паттерн проектирования, который состоит из трех основных компонентов:

Models (Модели) — работа с данными и БД.

Templates (Шаблоны) — отображение данных (HTML + логика).

Views (Представления) — обработка запросов и связь моделей с шаблонами.

3) Представление работает с моделью (данными, которые берутся из бд) и шаблонами (внешний вид/оформление информации, которая отдается клиенту)

Струткра приложения:



1) init.py -> приложения в Django формируются как пакет языка python    
2) migrations -> хранит миграции баз данных приложения
3) admin.py -> для настройки админ-панели нашего сайта
4) apps.py -> для настройки конфигурирования текущего приложения
5) models.py -> для хранения ORM-моделей - для взаимодействия с базой данных
6) test.py -> модель с тестирующими процедурами
7) views.py -> для хранения представлений текущего приложений

после создания приложения, его нужно зарегистрировать, чтоб фреймворк Django знал о его существовании, для этого:
1) открываем главную папку (в моем случае sitewomen)
2) переходим в файл settings.py (в этом файле уже прописано несколько стандартных приложений)
3) ниже просто добавляем свое, прописываем имя приложения в виде строки или желательно сразу добавляем путь к конфигурирущему файлу приложения (apps.py)

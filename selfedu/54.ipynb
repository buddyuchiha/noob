__#54. Класс FormView__

FormView наследуется от TemplateView и добавляет функциональность для:

1) Отображения формы (GET-запрос)

2) Обработки отправки формы (POST-запрос)

3) Валидации данных формы

4) Перенаправления после успешной обработки формы
